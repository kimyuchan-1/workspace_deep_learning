```python
from tqdm import tqdm
import time
import random
```

```python
items = [f'item_{i}' for i in range(30)]
for i in tqdm(range(len(items)), desc="데이터 처리 중"):
    time.sleep(0.01)
```

```python
for i in tqdm(items, desc="데이터 처리 중"):
    time.sleep(0.01)
```

```python
# 추가 정보 표시
pbar = tqdm(range(20), desc = "학습 중", colour="green")

for i in pbar:
    loss = random.uniform(0.1,10)
    acc = random.uniform(80,100)
    pbar.set_postfix({"loss":f"{loss:.4f}", "acc":f"{acc:.2f}"})
    time.sleep(0.1)
pbar.close()
```

```python
# 중첩 루프

outer_loop = tqdm(range(5), desc = "외부", colour="green")
for i in outer_loop:
    inner_loop = tqdm(range(10), desc = f"내부 {i+1}", colour="red", leave=False)
    for j in inner_loop:
        time.sleep(0.05)
    inner_loop.close()
outer_loop.close()
```
